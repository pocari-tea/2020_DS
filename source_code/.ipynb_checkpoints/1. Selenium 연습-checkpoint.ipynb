```python
from selenium import webdriver as wb
from selenium.webdriver.common.keys import Keys
from bs4 import BeautifulSoup as bs
import time
import pandas as pd
```

```python
#웹 브라우저 실행
driver = wb.Chrome("./chromedriver.exe")
```

```python
#해당 URL을 브라우저로 실행
url = 'https://www.naver.com'
driver.get(url)
```

```python
#검색창 태그(요소) 검색
input_search = driver.find_element_by_id('query')
```

```python
#검색창에 '더위는 언제 끝나나요?' 를 입력
input_search.send_keys('광주날씨')
```

```python
input_search.send_keys(Keys.ENTER)
```

```python
driver.back()
```

```python
#검색방법 2

#검색창 태그(요소) 검색
input_search = driver.find_element_by_id('query')

#검색창에 '더위는 언제 끝나나요?' 를 입력
input_search.send_keys('광주날씨')

#검색버튼 태그(요소) 검색
btn_search = driver.find_element_by_id('search_btn')

#검색버튼 클릭
btn_search.click()
```

```python
url = 'https://www.hsd.co.kr/menu/menu_list'

driver = wb.Chrome()
driver.get(url)
```

```python
onclick = driver.find_element_by_class_name('c_05')

for index in range(3):
    onclick.click()
    time.sleep(2)
```

```python
driver.page_source
```

```python
# 모든 도시락 정보가 나온 경우 bs4로 변환
soup = bs(driver.page_source, 'html.parser')
# lxml, html.parser
soup
```

```python
name_list = []
name_tag_list = soup.select("h4.h.fz_03")
for name_tag in name_tag_list:
    name_list.append(name_tag.text)
    print(name_tag.txt)
```

```python
name_list = []
name_tag_list = soup.select("h4.h.fz_03"):
for name_tag in soup.select("h4.h.fz_03"):
    name_list.append(name_tag.text)
name_list
```
