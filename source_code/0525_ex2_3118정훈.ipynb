```python
#제안 분야별 투표수
df2 = df.groupby(by = '제안 분야')[['투표 수']].sum()
df2 = df2.sort_values(by = '투표 수', ascending = False)
df2
```

```python
df3 = df.pivot_table(df, index = ['제안 분야'], aggfunc = 'sum')
df3
```

```python
#제목은 '제안분야별 투표수'
#X축 제목은 '투표수'
#그래프 색깔은 'yellow'
#격자무늬 넣기
#x축 범위 지정 (0~50까지)

df2['투표 수'].sort_values().plot(kind='barh', color='yellow', grid=True, figsize=(6, 6), title='제안분야별 투표수')
plt.xlabel('투표 수')
plt.xlim([0, 50])
plt.grid()
for i in range(3):
    plt.text(df2['투표 수'][i], 10-i, df2['투표 수'][i], fontsize=13)
plt.show()
```
