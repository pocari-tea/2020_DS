# 2. 데이터 수집

#### 사이킷런에 내장되어 있는 데이터(load_boston) 불러오고 그 데이터를 dataset 변수에 저장하기

# 3. 데이터 전처리

```python
import pandas as pd
```

#### data는 dataset.data, 칼럼명은 dataset.feature_names으로 데이터프레임을 만들어 bostonDF 에 저장하기 + 5줄 보이기

```python
data = pd.DataFrame(dataset.data, columns=dataset["feature_names"])
data.head()
```

#### bostonDF에 'PRICE'라는 새로운 칼럼 만들기, 이 안에는 정답(Lable - dataset.target)이 들어감 + 5줄 보이기

```python
data["PRICE"] = dataset.target
data.head()
```

#### 결측치 확인하기

```python
data.info()
```

#### bostonDF을 문제데이터(X), 정답데이터(y)로 분리하기  --> 분리한 이후 X.shape과 y.shape을 통해 잘 분리되었는지 확인해보기
- X는 bostonDF에서 'PRICE' 칼럼 삭제하여 저장
- y는 bostonDF에서 'PRICE' 칼럼만 저장

```python
X = data.iloc[:, :-1]
y = data["PRICE"]
```

```python
X.shape
```

```python
y.shape
```

# 5. 학습/ 테스트 데이터 나누기

#### test_size=0.3, random_state=0으로 지정하기

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=0)
```

```python
print(X_train.shape, y_train.shape)
print(X_test.shape, y_test.shape)
```

# 6. 모델 선택(LinearRegression), 학습

```python
from sklearn.linear_model import LinearRegression

model = LinearRegression()
```

```python
model.fit(X_train, y_train)
```

# 7. 모델의 계수, 절편 확인

```python
print("계수:{}".format(model.coef_))
print("절편:{}".format(model.intercept_))
```

# 8. 평가(R Squared, RMSE 둘 다 계산해보기)

```python
score = model.score(X_test, y_test)

pred = model.predict(X_test)
```

```python
import numpy as np
from sklearn.metrics import mean_squared_error, r2_score

r2_score(y_test, pred)
```

```python
np.sqrt(mean_squared_error(y_test,pred))  # rmse
```

# 9. 특성 늘리기, MinMaxScaler 하여 다시 학습, 평가하기

```python
col = X.columns
for i in range(col.size):
    for j in range(i,col.size):
        X[col[i]+'x'+col[j]] = X[col[i]]*X[col[j]]
```

```python
print(X.shape)
X.head()
```

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
scaler = MinMaxScaler()
X = scaler.fit_transform(X)
```

```python
X = pd.DataFrame(X)
X.head()
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.3, random_state=1)
```

lr2 = LinearRegression()

# 10. Ridge 사용하여 과대적합 해결하기

```python
from sklearn.linear_model import Ridge, Lasso
```

```python
model = Ridge(alpha=0.025)
model.fit(X_train, y_train)

print("훈련 세트 점수:", model.score(X_train, y_train))
print("테스트 세트 점수:", model.score(X_test, y_test))
```

```python
pred = model.predict(X_test)
np.sqrt(mean_squared_error(y_test,pred))  # rmse
```

# 11. Lasso 사용하여 과대적합 해결하기

```python
model = Lasso(alpha=0.001)
model.fit(X_train, y_train)

print("훈련 세트 점수:", model.score(X_train, y_train))
print("테스트 세트 점수:", model.score(X_test, y_test))
```

```python
pred = model.predict(X_test)
np.sqrt(mean_squared_error(y_test,pred))  # rmse
```

# 12. KNN 회귀

```python
from sklearn.neighbors import KNeighborsRegressor
```

```python
model = KNeighborsRegressor(n_neighbors=2)
model.fit(X_train, y_train)

print("훈련 세트 점수:", model.score(X_train, y_train))
print("테스트 세트 점수:", model.score(X_test, y_test))
```

```python
pred = model.predict(X_test)
np.sqrt(mean_squared_error(y_test,pred))  # rmse
```
